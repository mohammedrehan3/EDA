```python
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first.head()
```

```python
bowling_first=ipl[ipl['win_by_wickets']!=0]
bowling_first.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(batting_first['win_by_runs'],color='gold',edgecolor='grey')
plt.title('Winning by runs')
plt.xlabel('Runs')
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),batting_first['winner'].value_counts()[0:5],color=['green','blue','red','yellow','grey'])
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.pie(list(batting_first['winner'].value_counts()[0:7]),labels=list(batting_first['winner'].value_counts()[0:7].keys()),autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.hist(bowling_first['win_by_wickets'],bins=20,color='g',edgecolor='red')
plt.xlabel('Wickets')
plt.show()
```

```python
bowling_first['winner'].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.bar(list(bowling_first['winner'].value_counts()[0:5].keys()),bowling_first['winner'].value_counts()[0:5],color=['green','blue','red','yellow','grey'])
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.pie(list(bowling_first['winner'].value_counts()[0:7]),labels=list(bowling_first['winner'].value_counts()[0:7].keys()),autopct='%1.1f%%')
plt.show()
```
